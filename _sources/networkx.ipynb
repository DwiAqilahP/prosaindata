```python
import networkx as nx
G=nx.Graph()
```

```python
G.add_edges_from([(1,2),(1,3),(1,4),(2,3),(2,5),(3,5),(4,5),(4,6),(5,6),(5,7),(6,7),(6,8),(7,8),(7,9),(8,9)])
```

```python
nx.draw(G,with_labels=True)
```

```python
Dc = nx.degree_centrality(G)
Dc
```

```python
sorted(Dc, key=Dc.get, reverse=True)
```

```python
cc = nx.closeness_centrality(G)
cc
```

```python
sorted(cc, key=cc.get, reverse=True)
```

```python
cb = nx.betweenness_centrality(G)
cb
```

```python
sorted(cb, key=cb.get, reverse=True)
```
